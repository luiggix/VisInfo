```python
fig, (ax1, ax2) = plt.subplots(2, 1)
# make a little extra space between the subplots
fig.subplots_adjust(hspace=0.5)

dt = 0.01
t = np.arange(0, 30, dt)

# Fixing random state for reproducibility
np.random.seed(19680801)


nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2
r = np.exp(-t / 0.05)

cnse1 = np.convolve(nse1, r, mode='same') * dt   # colored noise 1
cnse2 = np.convolve(nse2, r, mode='same') * dt   # colored noise 2

# two signals with a coherent part and a random part
s1 = 0.01 * np.sin(2 * np.pi * 10 * t) + cnse1
s2 = 0.01 * np.sin(2 * np.pi * 10 * t) + cnse2

ax1.plot(t, s1, t, s2)
ax1.set_xlim(0, 5)
ax1.set_xlabel('time')
ax1.set_ylabel('s1 and s2')
ax1.grid(True)

cxy, f = ax2.csd(s1, s2, 256, 1. / dt)
ax2.set_ylabel('CSD (db)')
plt.savefig('./Imagenes/info2D.png')
```

```python
from matplotlib.patches import Circle

def E(q, r0, x, y):
    """
    Calcula el campo elctrico vectorial generado por una 
    carga q en el punto r0.
    
    Args:
        q : carga elctrica
        r0: tupla que indica la posicin de la carga elctrica.
        x,y: grid para calcular el campo elctrico.
    Return:
        El campo elctrico vectorial E=(Ex,Ey)
    """
    r3 = np.hypot(x-r0[0], y-r0[1])**3
    return q * (x - r0[0]) / r3, q * (y - r0[1]) / r3

def generaCargas(nq):
    """
    Crea multipolos con un nmero de cargas nq de signo 
    alternante igualmente espaciadas en el crculo 
    unitario.
    
    Args:
        nq : nmero total de cargas.

    Return:
    """
    cg = []
    for i in range(nq):
        q = i%2 * 2 - 1
        x = np.cos(2*np.pi*i/nq)
        y = np.sin(2*np.pi*i/nq)
        cg.append((q,(x,y)))
    return cg

# Creacin del dominio y la malla
nx, ny = 32, 32
lx1,lx2 = -4, 4
ly1,ly2 = -2, 2
xg,yg = np.meshgrid(np.linspace(lx1, lx2, nx), np.linspace(ly1, ly2, ny))

nq = 2**2
cargas = generaCargas(nq)

# Crea el campo vectorial E = (Ex,Ey) sumando los campos
# elctricos que genera cada carga.
Ex, Ey = np.zeros((ny, nx)), np.zeros((ny, nx))
for carga in cargas:
    ex, ey = E(*carga, x=xg, y=yg)
    Ex += ex
    Ey += ey

#ax = fig.add_subplot(111)

fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(10,6))

# Mapa de color para dibujar las lneas de corriente.
mapa_colores = 2 * np.log(np.hypot(Ex, Ey))

ax[0].streamplot(xg, yg, Ex, Ey, color=mapa_colores, linewidth=0.5, 
              cmap=plt.cm.inferno, density=3, arrowstyle='->', 
              arrowsize=1.5)

# Agregamos crculos en las posiciones de las cargas
colores = {True: '#aa0000', False: '#0000aa'}
for q, pos in cargas:
    ax[0].add_artist(Circle(pos, 0.05, color=colores[q>0]))

ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$y$')
ax[0].set_xlim(lx1,lx2)
ax[0].set_ylim(ly1,ly2)
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_aspect('equal')

w = 3
Y, X = np.mgrid[-2:2:100j, -4:4:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U**2 + V**2)

lw = 5*speed / speed.max()
ax[1].streamplot(X, Y, U, V, density=0.6, color='k', linewidth=lw)
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$y$')
ax[1].set_xlim(lx1,lx2)
ax[1].set_ylim(ly1,ly2)
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_aspect('equal')
#plt.title('Campo Elctrico $E = (E_x, E_y)$')
#plt.savefig('Electrico.pdf')

fig.tight_layout()

plt.savefig('./Imagenes/campo1.png')
```

```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import


def explode(data):
    size = np.array(data.shape)*2
    data_e = np.zeros(size - 1, dtype=data.dtype)
    data_e[::2, ::2, ::2] = data
    return data_e

# build up the numpy logo
n_voxels = np.zeros((4, 3, 4), dtype=bool)
n_voxels[0, 0, :] = True
n_voxels[-1, 0, :] = True
n_voxels[1, 0, 2] = True
n_voxels[2, 0, 1] = True
facecolors = np.where(n_voxels, '#FFD65DC0', '#7A88CCC0')
edgecolors = np.where(n_voxels, '#BFAB6E', '#7D84A6')
filled = np.ones(n_voxels.shape)

# upscale the above voxel image, leaving gaps
filled_2 = explode(filled)
fcolors_2 = explode(facecolors)
ecolors_2 = explode(edgecolors)

# Shrink the gaps
x, y, z = np.indices(np.array(filled_2.shape) + 1).astype(float) // 2
x[0::2, :, :] += 0.05
y[:, 0::2, :] += 0.05
z[:, :, 0::2] += 0.05
x[1::2, :, :] += 0.95
y[:, 1::2, :] += 0.95
z[:, :, 1::2] += 0.95

fig = plt.figure(figsize=(10,8))
ax = fig.gca(projection='3d')
ax.voxels(x, y, z, filled_2, facecolors=fcolors_2, edgecolors=ecolors_2)

plt.savefig('./Imagenes/volume.png')
```
