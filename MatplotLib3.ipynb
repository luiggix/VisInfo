```python
import pandas as pd
from pandas import DataFrame
```

```python
dframe = pd.read_csv('Estado_de_Situación_Financiera.csv').dropna()
dframe
fig = plt.figure()
plt.scatter( [i for i in range(0,len(dframe['Monto']))], 
             dframe['Monto'])
plt.ylabel('Monto [\$]')
```

```python
x = [i for i in range(0,len(dframe['Monto']))]
y = dframe['Monto']
fig, axes = plt.subplots(2,2)
plt.tight_layout()
axes[1,0].set_xlabel('eje x', color='red')
axes[1,1].set_xlabel('eje x', color='red')
axes[0,0].set_ylabel('eje y', color='blue')
axes[1,0].set_ylabel('eje y', color='blue')
axes[0,0].spines['right'].set_visible(False)
axes[0,0].spines['top'].set_visible(False)

axes[0,0].plot(x,y, 'r--')
axes[0,1].scatter(x,y)
axes[1,0].plot(x,np.sin(x),c='green')
axes[1,1].plot(x,np.exp(x))
```
